```python
import numpy as np 
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,21)
x
```

```python
y = [2,  4,  6,  8, 2, 12, 5, 16, 25, 30, 22, 24, 26, 28, 30, 28, 30,
       32, 35, 40]
y
```

```python
# plt.plot(x,y) # without dotts
plt.plot(x,y, marker=".") # with dotts
# plt.style.use("ggplot") # chart styling
plt.savefig("plot.jpg") # saving image of plot
plt.show()
```

```python
plt.plot(x,y, color="g", linestyle=":", linewidth=5)
plt.title("Line plot X vs Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(["X"])
plt.show()
```

```python
y1, y2 = x*2, x*3
plt.plot(x, y1, color="g", linestyle=":", linewidth=2)
plt.plot(x, y2, color="r", linestyle="-", linewidth=3)
plt.title("two Lines in a plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(["X","Y"])
plt.tight_layout() 
# plt.xkcd() # comic design
plt.grid()
plt.show()
```

```python
plt.subplot(1,2,1) # 1 row 2 column 1 index
plt.plot(x, y1, color='b' , linestyle=':', linewidth=2)

plt.subplot(1,2,2) # 
plt.plot(x, y2, color='r', linestyle=':', linewidth=3)
plt.title
plt.show()
```

```python
plt.subplot(2,1,1) # row, column, index
plt.plot(x, y1, color='b' , linestyle=':', linewidth=2)
plt.subplot(2,1,2) # 
plt.plot(x, y2, color='r', linestyle=':', linewidth=3)
plt.title
plt.show()
```

```python
student = {"Bob":45, "Matt":55, "Tony":75,"Berlin": 70}
name = list(student.keys())  
marks = student.values()
```

```python
plt.bar(name,marks)  # vertical Bar
plt.title("Students Marks")
plt.xlabel("Names of Students")
plt.ylabel("Marks of Students")
# plt.grid()
plt.show()
```

```python
plt.barh(name, marks, color="g")
plt.title("Students Result")
plt.xlabel("Marks of Students")
plt.ylabel("Names of Students")
plt.show()
```

```python
fruits = ["Apple", "Orange", "Mango","Banana"]
quantity = [15, 20, 40, 70]
# plt.pie(quantity, labels=fruits) # simple
plt.pie(quantity, labels=fruits, autopct='%0.1f%%', colors=["Red", "Orange", "Yellow", "green"]) # modified 
plt.title("Fruits Stored")
plt.show()
```

```python
fruits = ["Apple", "Orange", "Mango","Banana"]
quantity = [15, 20, 40, 70] 
plt.pie(quantity, labels=fruits, radius=2) # radius is size
plt.pie([1], colors=['w'], radius=1) 
plt.show()
```
