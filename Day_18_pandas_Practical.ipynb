```python
country_use_python = country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains("Python").sum())
country_use_python
```

```python
python_df = pd.concat([country_resp, country_use_python],axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'Country':'NumRespondents', 'LanguageHaveWorkedWith':'NumKnowsPython'},inplace=True)
python_df
```

```python
python_df['perctKnows_Python'] = (python_df["NumRespondents"]/python_df["NumKnowsPython"])*100
python_df
```

```python
python_df.sort_values(by='perctKnows_Python')
```

# Cleaning Data - Casting Datatypes and Handling Missing Values

```python
import numpy as np
na_values = ['NA', 'Missing']
df = pd.read_csv('files_panda/survey_results_public.csv', na_values=na_values)
df.head()
```

```python
df['YearsCode'].value_counts()
```

```python
df['YearsCode'].unique()
```

```python
df['YearsCode'].replace({'Less than 1 year':0, 'More than 50 years': 51},inplace=True)
df['YearsCode'].unique()
```

```python
df['YearsCode'] = df["YearsCode"].astype(float)
```

```python
df["YearsCode"].mean()
```

```python
df["YearsCode"].median()
```
